```python
import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
```

1.Start by performing a GET request on the url above and convert the response into a BeautifulSoup object.

```python
URL = 'https://realpython.github.io/fake-jobs/'

response = requests.get(URL)
```

```python
response.status_code
```

```python
soup = BS(response.text)
```

```python
print(soup.prettify())
```

1b. Now, use what you did for the first title, but extract the job title for all jobs on this page. Store the results in a list.

```python
job_titles = soup.findAll('h2')
```

```python
 job_titles_list = [title.text for title in job_titles]
```

1c. Finally, extract the companies, locations, and posting dates for each job.

```python
company_titles = soup.findAll('h3')
```

```python
 companies_list = [company.text for company in company_titles]
```

```python
locations = soup.findAll('p', attrs={'class' : 'location'})
```

```python
 locations_list = [loc.text.strip() for loc in locations]
```

```python
time = soup.findAll('time', attrs={'datetime' : '2021-04-08'})
```

```python
 date_list = [date.text.strip() for date in time]
```

3. Finally, we want to get the job description text for each job.

```python
def get_job_description(job_url):
    soup = BS(requests.get(job_url).text, 'html.parser')
    return soup.find("div", class_="content").p.text.strip()
```

```python
get_job_description('https://realpython.github.io/fake-jobs/jobs/senior-python-developer-0.html')
```

```python
jobs['Description'] = jobs['Apply URL'].apply(get_job_description)
```
